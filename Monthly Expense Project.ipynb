```python
import matplotlib.pyplot as plt

expense_types = [
    "Housing",
    "Mortgage & Debts Payments",
    "Utilities",
    "Healthcare",
    "Groceries",
    "Dining Out",
    "Savings & Investments",
    "Transportations",
    "Entertainment & Gift",
    "Others"
]

# Collect expenses from user
expenses = []
print("Please enter your expenses for each category:")

for expense in expense_types:
    while True:
        try:
            value = float(input(f"{expense}: "))
            if value < 0:
                print("Please enter a non-negative number.")
                continue
            expenses.append(value)
            break
        except ValueError:
            print("Invalid input. Please enter a number.")


total = sum(expenses)
percentages = [100 * x / total for x in expenses]

plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=expense_types, autopct='%1.1f%%', startangle=140)
plt.title("Monthly Expense Breakdown")
plt.axis('equal')
plt.show()
```
